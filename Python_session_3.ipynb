```python
# to find, if a number is prime or not ?

number = int(input("Please enter your number : "))

#  either divisible by 1 or divisble by itself

isPrime = True

for n in range(2, number):
  if(number%n ==0):
    isPrime = False
    break


if isPrime == True:
  print("Number is prime : ", number)
else:
  print("Number is not a prime number : ", number)
```
